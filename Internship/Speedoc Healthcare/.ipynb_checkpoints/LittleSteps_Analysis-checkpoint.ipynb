### 2.2 Data Cleaning

Apply five cleaning steps in a deliberate order: duplicates → datetimes → drop missing end times → categorical corrections → missing notes. This sequence ensures we do not redundantly parse datetimes on rows that will be dropped.


````python
# Step 1 — Remove duplicate visit_id records (keep first occurrence)
# Justification: visit_id uniquely identifies an encounter.
# Duplicates are data entry errors; we keep the first (original) record.

n_before = len(df_raw)
df = df_raw.drop_duplicates(subset=["visit_id"], keep="first").reset_index(drop=True)
print(f"Rows before: {n_before:,}  |  After: {len(df):,}  |  Removed: {n_before - len(df)}")
````

````python
# Step 2 — Standardise datetime columns
# Five formats are present; dateutil.parser auto-detects all of them.
# We wrap in try/except to return None (not raise) on unparseable values.

def safe_parse_dt(value):
    """Parse any datetime string format present in the dataset."""
    if pd.isna(value):
        return None
    try:
        return dtparser.parse(str(value).strip())
    except (ValueError, OverflowError):
        return None

for col in ["visit_start_time", "visit_end_time"]:
    df[col] = pd.to_datetime(df[col].apply(safe_parse_dt))

print("visit_start_time dtype:", df["visit_start_time"].dtype)
print("visit_end_time dtype  :", df["visit_end_time"].dtype)
print("Missing visit_end_time after parse:", df["visit_end_time"].isna().sum())
````

````python
# Step 3 — Drop rows with missing visit_end_time
# Justification: visit_end_time is required for duration calculation.
# Imputing end times with the median would compress variance and bias
# all downstream analyses — dropping is the more honest choice.

n_before = len(df)
df = df.dropna(subset=["visit_end_time"]).reset_index(drop=True)
print(f"Rows dropped (missing visit_end_time): {n_before - len(df)}")
print(f"Remaining rows: {len(df):,}")
````

````python
# Step 4 — Correct categorical typos
# Typos were identified during exploration by inspecting all unique values.
# Deterministic string replacement is used (not fuzzy matching) because
# all correct canonical forms are unambiguous.

SERVICE_TYPE_MAP = {
    "Pyhcisal Therapy":       "Physical Therapy",
    "Wound Cae":              "Wound Care",
    "General Chek-up":        "General Check-up",
    "Medicatn Adminstratino": "Medication Administration",
}
LOCATION_MAP = {
    "Notrh":  "North",
    "Easst":  "East",
    "Wsst":   "West",
    "Soutth": "South",
}

df["service_type"]   = df["service_type"].replace(SERVICE_TYPE_MAP)
df["visit_location"] = df["visit_location"].replace(LOCATION_MAP)

print("service_type after correction:",   sorted(df["service_type"].unique()))
print("visit_location after correction:", sorted(df["visit_location"].unique()))
````

````python
# Step 5 — Handle missing nurse_notes
# Notes are supplementary. Rows with missing notes are retained for duration
# analysis and filled with a sentinel string so they are identifiable and
# excluded from text analysis without dropping the entire row.

n_missing_notes = df["nurse_notes"].isna().sum()
df["nurse_notes"] = df["nurse_notes"].fillna("No notes provided")
print(f"nurse_notes: {n_missing_notes} missing values filled with 'No notes provided'")
````

````python
# Cleaning summary
print("=" * 52)
print(f"  Raw dataset rows:                    {len(df_raw):>5,}")
print(f"  After deduplication (-30 dupes):     {len(df_raw)-30:>5,}")
print(f"  After dropping no-end-time rows:     {len(df):>5,}")
print(f"  Columns with standardised types:     visit_start_time, visit_end_time")
print(f"  Columns with corrected categories:   service_type, visit_location")
print(f"  Columns with imputed text:           nurse_notes")
print("=" * 52)
````

## 3. Part 2 — Feature Engineering

### 3.1 Visit Duration

`visit_duration_minutes` is derived directly from the cleaned, standardised timestamps. This is the primary metric for all downstream analysis.


````python
df["visit_duration_minutes"] = (
    (df["visit_end_time"] - df["visit_start_time"]).dt.total_seconds() / 60
)

print("visit_duration_minutes — raw range (before outlier removal):")
print(df["visit_duration_minutes"].describe().round(2))
````

### 3.2 Outlier Removal

The raw durations span **0.4 to 1,432 minutes** (~24 hours). Values at the extremes represent clear data entry errors — a 24-hour "visit" is clinically implausible for at-home healthcare.

**Method: IQR fence (1.5×)**

$$\text{Lower fence} = Q_1 - 1.5 \times IQR \qquad \text{Upper fence} = Q_3 + 1.5 \times IQR$$

This is the standard statistical approach for skewed distributions. It preserves legitimately long visits (e.g., a complex 2-hour wound care session) while removing physically impossible values caused by timestamp errors.


````python
# IQR outlier removal on visit_duration_minutes
df_before_outliers = df.copy()   # retained for histogram comparison

q1 = df["visit_duration_minutes"].quantile(0.25)
q3 = df["visit_duration_minutes"].quantile(0.75)
iqr = q3 - q1
lower_fence = q1 - 1.5 * iqr
upper_fence = q3 + 1.5 * iqr

n_before = len(df)
df = df[
    (df["visit_duration_minutes"] >= lower_fence) &
    (df["visit_duration_minutes"] <= upper_fence)
].copy().reset_index(drop=True)

print(f"IQR fence: [{lower_fence:.1f}, {upper_fence:.1f}] minutes")
print(f"Rows removed: {n_before - len(df)}  ({(n_before-len(df))/n_before*100:.1f}%)")
print(f"Rows remaining: {len(df):,}")
````

### 3.3 Travel Duration

> **Assumption — documented explicitly**
>
> The dataset contains **no direct travel time measurements** (no GPS, departure/arrival timestamps, or odometer readings).
>
> We derive a proxy: for each nurse, visits are sorted chronologically and the **inter-visit gap** (time between one visit ending and the next beginning) is computed. This is an upper bound — it includes travel time plus any idle time between visits.
>
> **Exclusions applied:**
> - First visit per nurse → NaN (no preceding visit to measure gap from)
> - Negative gaps (timestamp error) → NaN
> - Gaps > 240 minutes (4 hours) → NaN (treated as a shift break, not travel)

This proxy is the most defensible approach given the available data, and is documented here and in the README.


````python
# Travel duration — inter-visit gap per nurse
df = df.sort_values(["nurse_id", "visit_start_time"]).reset_index(drop=True)

# Previous visit's end time for each nurse (shift by 1 within each nurse group)
df["_prev_end"] = df.groupby("nurse_id")["visit_end_time"].shift(1)

# Gap = current visit start - previous visit end (in minutes)
df["travel_duration_minutes"] = (
    (df["visit_start_time"] - df["_prev_end"]).dt.total_seconds() / 60
)

# Nullify invalid / implausible gaps
df.loc[df["travel_duration_minutes"] < 0,   "travel_duration_minutes"] = np.nan  # timestamp error
df.loc[df["travel_duration_minutes"] > 240, "travel_duration_minutes"] = np.nan  # shift break

df = df.drop(columns=["_prev_end"])

n_valid = df["travel_duration_minutes"].notna().sum()
print(f"Valid travel_duration_minutes values: {n_valid} / {len(df)}")
print(f"(Most nurses appear only once in the dataset, yielding no inter-visit gap.)")
print()
print(df["travel_duration_minutes"].describe().round(2))
````

### 3.4 Nurse Notes — Structured Feature Extraction

The `nurse_notes` field follows a semi-structured natural language pattern:

```
[Urgency phrase.] The nurse [action] the [subject]. The patient is [status].
```

We use **regular expressions** to extract three structured features:

| Feature | What it captures | Example values |
|---|---|---|
| `urgency_flag` | Clinical urgency level | ASAP, Critical, Urgent, Review, Follow-Up, None |
| `patient_status` | Patient condition at time of visit | Stable, In Pain, Dizzy, Weak, Restless, Comfortable, Improving |
| `action_taken` | Primary nursing action performed | `reviewed the blood pressure` |


````python
URGENCY_PATTERN = re.compile(
    r"\b(ASAP|Critical|Urgent|Action|Review|Monitoring|Ongoing assessment|Follow-up)\b",
    re.IGNORECASE,
)
STATUS_PATTERN = re.compile(
    r"The patient is (stable|in pain|dizzy|weak|restless|comfortable|improving)",
    re.IGNORECASE,
)
ACTION_PATTERN = re.compile(r"The nurse (\w+) the ([^.]+)\.", re.IGNORECASE)

def extract_urgency(note):
    if note == "No notes provided":
        return "None"
    matches = URGENCY_PATTERN.findall(note)
    return matches[0].title() if matches else "None"

def extract_status(note):
    m = STATUS_PATTERN.search(note)
    return m.group(1).title() if m else "Unknown"

def extract_action(note):
    m = ACTION_PATTERN.search(note)
    return f"{m.group(1).lower()} the {m.group(2).strip()}" if m else "Unknown"

df["urgency_flag"]   = df["nurse_notes"].apply(extract_urgency)
df["patient_status"] = df["nurse_notes"].apply(extract_status)
df["action_taken"]   = df["nurse_notes"].apply(extract_action)

print("Extraction complete. Sample results:")
df[["nurse_notes", "urgency_flag", "patient_status", "action_taken"]].head(6)
````

````python
order = (df.groupby("service_type")["visit_duration_minutes"]
           .median().sort_values(ascending=False).index.tolist())

fig, ax = plt.subplots(figsize=(10, 5))
sns.boxplot(data=df, x="service_type", y="visit_duration_minutes",
            order=order, palette="Blues", ax=ax,
            flierprops=dict(marker="o", markersize=3, alpha=0.4))
ax.set_xlabel("Service Type")
ax.set_ylabel("Visit Duration (minutes)")
ax.set_title("Visit Duration Distribution by Service Type", fontweight="bold")
plt.xticks(rotation=18, ha="right")
plt.tight_layout()
plt.show()
````

**Analysis:**

The mean differences across service types are modest in absolute terms (~4 minutes between Physical Therapy and General Check-up). However, the boxplots reveal important nuances:

- **Physical Therapy** has the longest median and the widest interquartile range. Therapy sessions vary considerably — some focus on simple exercises, others on intensive rehabilitation — explaining the spread.
- **General Check-up** visits are the shortest on average but still highly variable. A check-up may range from a quick vital-signs reading to a comprehensive assessment.
- All four service types show **similar distributional shapes**, indicating that within-type variability driven by patient complexity substantially exceeds between-type variability.

**Conclusion:** Service type alone is a weak predictor of visit duration. Patient complexity (captured partly in nurse notes) is a stronger driver. Nonetheless, even a 4-minute difference per visit, compounding across 8 visits per shift, adds ~32 minutes of unaccounted schedule drift — enough to justify service-specific time blocks.

### 5.3 Visit Duration by Location Zone


````python
# Kruskal-Wallis test — non-parametric, does not assume equal variances
# Chosen over one-way ANOVA because within-group normality is not guaranteed
# across all zones and the sample sizes differ slightly.

groups = [
    g["visit_duration_minutes"].dropna().values
    for _, g in df.groupby("visit_location")
]
h_stat, p_val = stats.kruskal(*groups)

print(f"Kruskal-Wallis H-statistic : {h_stat:.4f}")
print(f"p-value                    : {p_val:.4f}")
print()
if p_val < 0.05:
    print("SIGNIFICANT: visit durations differ across zones (p < 0.05)")
else:
    print("NOT significant: no reliable zone effect detected (p >= 0.05)")

print()
print("Average visit duration by zone:")
print(df.groupby("visit_location")["visit_duration_minutes"].mean().sort_values(ascending=False).round(2))
````

````python
loc_avg = (
    df.groupby("visit_location")["visit_duration_minutes"]
    .mean().sort_values(ascending=False).reset_index()
)
loc_avg.columns = ["zone", "avg"]
overall_mean = df["visit_duration_minutes"].mean()

fig, ax = plt.subplots(figsize=(7, 4))
colors = [WARN_COLOR if v > overall_mean else ACCENT for v in loc_avg["avg"]]
bars = ax.bar(loc_avg["zone"], loc_avg["avg"], color=colors, edgecolor="white", width=0.6)
ax.axhline(overall_mean, color="black", linestyle="--", linewidth=1.2,
           label=f"Overall mean: {overall_mean:.1f} min")

for bar, val in zip(bars, loc_avg["avg"]):
    ax.text(bar.get_x() + bar.get_width() / 2, val + 0.4,
            f"{val:.1f}", ha="center", va="bottom", fontsize=9)

ax.set_xlabel("Visit Location Zone")
ax.set_ylabel("Average Visit Duration (minutes)")
ax.set_title("Average Visit Duration by Location Zone", fontweight="bold")
ax.legend()
plt.tight_layout()
plt.show()
````

````python
fig, ax = plt.subplots(figsize=(8, 4))
sns.boxplot(data=df, x="visit_location", y="visit_duration_minutes",
            palette="Set2", ax=ax,
            flierprops=dict(marker="o", markersize=3, alpha=0.4))
ax.set_xlabel("Visit Location Zone")
ax.set_ylabel("Visit Duration (minutes)")
ax.set_title("Visit Duration Distribution by Location Zone", fontweight="bold")
plt.tight_layout()
plt.show()
````

**Analysis:**

The Kruskal-Wallis test returns **H = 8.41, p = 0.038**, confirming a **statistically significant difference** in visit durations across zones at the 5% level.

- **North** (65.9 min) and **South** (64.9 min) average above the overall mean of 62.6 min (red bars).
- **West** zone (57.3 min) is ~8.6 minutes shorter than North — a meaningful operational difference.

The boxplots show comparable within-zone variance, so the zone effect is not driven by a handful of extreme visits; it is a genuine distributional shift.

**Possible explanations:**
1. North/South zones may serve a higher proportion of complex chronic-disease patients.
2. Housing characteristics (multi-storey buildings, secured access) in northern zones may add non-clinical time to each visit.
3. Nurse assignment may inadvertently cluster less experienced nurses in the West, leading to faster but potentially less thorough visits.

A case-mix audit by zone is recommended before drawing firm operational conclusions.

````python
# Scatter — nurse-level travel vs visit duration
nurse_summary = df.groupby("nurse_id").agg(
    avg_visit  = ("visit_duration_minutes",  "mean"),
    avg_travel = ("travel_duration_minutes", "mean"),
    n_visits   = ("visit_id",                "count"),
).dropna()

fig, ax = plt.subplots(figsize=(7, 5))
sc = ax.scatter(
    nurse_summary["avg_travel"], nurse_summary["avg_visit"],
    s=nurse_summary["n_visits"] * 3,
    c=nurse_summary["n_visits"], cmap="Blues",
    edgecolors="white", linewidths=0.5, alpha=0.85,
)
plt.colorbar(sc, ax=ax, label="Number of visits")
ax.set_xlabel("Avg Travel Duration (minutes)")
ax.set_ylabel("Avg Visit Duration (minutes)")
ax.set_title("Nurse-Level: Travel Duration vs Visit Duration\n(bubble size = visit count)",
             fontweight="bold")
plt.tight_layout()
plt.show()
````

**Analysis:**

The gap between the highest and lowest-travel nurses is an 18× difference: nurse **N9783** averages **225.9 minutes** between visits versus nurse **N8068** at just **12.1 minutes**. Even accounting for the fact that our travel proxy includes idle time, this is an extreme operational disparity.

The scatter plot shows **no meaningful relationship** between average travel time and average visit duration — nurses with long travel times are not compensating by rushing through their visits. This is reassuring from a quality-of-care perspective, but confirms the travel overhead is pure operational waste, not a quality-complexity tradeoff.

**Important caveat:** Only 34 nurses have consecutive visits in this dataset, limiting the generalisability of individual nurse rankings. A larger multi-week dataset would be needed to reliably identify structurally high-travel nurses versus those who happened to have one long inter-visit gap in this extract.

### 5.5 Insights from Nurse Notes

**Analysis — Nurse Notes Insights:**

**Urgency flags and visit duration:**
- **Urgent** (69.4 min) and **ASAP** (65.8 min) flags are associated with the longest visits — approximately 7–9 minutes above the mean. These cases require additional assessment, intervention, or stabilisation before the nurse can safely depart.
- **Critical** visits (60.6 min) are shorter than expected — a counterintuitive finding. One plausible explanation: "Critical" situations trigger rapid handoff to emergency services, so the nurse's on-site time is actually brief (stabilise → call → hand off), rather than managing the full visit autonomously.
- **Follow-Up** visits (59.1 min) are the shortest, consistent with their nature as structured, planned re-assessments rather than acute interventions.

**Patient status and visit duration:**
- **Restless** (69.5 min) and **Dizzy** (65.2 min) patients require the most time. These conditions demand careful safety assessment and monitoring before the nurse can safely conclude the visit.
- **In Pain** (57.6 min) visits are below average — likely because effective targeted interventions (analgesia, positioning) resolve the immediate issue quickly, allowing the nurse to depart efficiently.
- **Stable** and **Comfortable** patients cluster near the mean, consistent with routine care.

**Text analysis conclusion:** Nurse notes contain statistically meaningful signal for predicting visit duration. An urgency/status-aware scheduling system — even a simple heuristic adding 10 minutes for Urgent/ASAP flags — would improve schedule accuracy over the current service-type-only approach.

## 6. Part 5 — Conclusions & Operational Recommendations

### Summary of Key Findings

| Finding | Detail |
|---|---|
| Average visit duration | **62.6 min** (after outlier cleaning, 849 records) |
| Average travel duration | **129 min** inter-visit gap proxy (34 valid records) |
| Longest service type | Physical Therapy (63.7 min avg) |
| Shortest service type | General Check-up (59.9 min avg) |
| Location zone effect | **Statistically significant** (Kruskal-Wallis p = 0.038) |
| Longest-zone | North (65.9 min) — 8.6 min above West |
| Shortest-zone | West (57.3 min) |
| Highest-travel nurse | N9783 (225.9 min avg inter-visit gap) |
| Lowest-travel nurse | N8068 (12.1 min avg inter-visit gap) |
| Urgency — longest visits | Urgent-flagged: +6.8 min above mean |
| Status — longest visits | Restless patients: +6.9 min above mean |

---

### Recommendations

**1. Service-type scheduling blocks**
Physical Therapy and Wound Care consistently run 3–4 min longer than Check-ups. Across 8 visits/shift this is ~32 min of unaccounted drift. Implement service-specific time blocks (e.g., 55 min for Check-ups, 65 min for Physical Therapy/Wound Care/Medication Administration).

**2. Geographic clustering for high-travel nurses**
Nurses N9783, N6322, and N7802 have inter-visit gaps nearly 70% above the fleet average. Clustering their patient assignments geographically — using postcode or zone-based batching — could recover 1–2 hours of productive clinical time per nurse per shift.

**3. Urgency-aware scheduling buffers**
Visits with Urgent or ASAP notes run ~7–9 min over mean duration. Build an automatic 10–15 min buffer into the next scheduled slot when a patient's recent notes carry these flags, to absorb overruns without cascading delays.

**4. North zone audit**
The North zone's statistically elevated duration warrants a follow-up audit covering patient case-mix complexity, building access issues, and nurse experience distribution.

**5. Field data quality improvements**
10% missing end timestamps and 3% duplicate records, both likely caused by manual field timestamping. Recommended fix: implement automatic timestamp capture via GPS geofencing in the nurse mobile app (auto-start on arrival, auto-close on departure).

**6. Predictive scheduling model (future work)**
Nurse notes contain significant predictive signal. A lightweight model — even linear regression on encoded urgency + status + service type — could generate per-visit duration estimates more accurate than the current uniform-block approach, enabling dynamic real-time schedule optimisation.

---

### Data Limitations & Assumptions

| Assumption | Rationale |
|---|---|
| Rows with missing `visit_end_time` dropped (not imputed) | Imputation would bias all duration statistics |
| Travel duration = inter-visit gap (proxy) | No GPS or departure/arrival data available; gap is an upper-bound proxy |
| IQR 1.5× outlier removal | Standard method; clinically defensible ceiling ~122 min |
| Typo correction is deterministic | Fuzzy matching risks false positives; exact match is safe given small, unambiguous typo set |
| Gaps > 240 min treated as shift breaks | At-home healthcare shifts rarely require cross-city transit exceeding 4 hours |
